```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('data.csv', header=None)
df
```

```python
plt.scatter(df[0],df[1])
```

#  The data is 2 dimensional. Increasing the dimension of our data using Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
```

```python
X=df[0].values
y=df[1].values
```

```python
X=X.reshape((X.shape[0],1))
X=poly.fit_transform(X)
```

```python
X=X/X.max()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=100)
```

```python
from Linear_Regression import LinearRegressionMGD
```

```python
model=LinearRegressionMGD()
```

```python
%time model.fit(X_train,y_train, lr=1e-2, batch_size=10, epochs=3000, threshold=1e-8)
```

# Loss Decreasing Graph

```python
plt.plot(range(len(model.costs)), model.costs)
plt.title('Loss at epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

# r2 score calculation

```python
from sklearn.metrics import r2_score
y_pred=model.predict(X_test)
r2_score(y_test,y_pred)
```
